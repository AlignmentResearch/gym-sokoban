```python
import numpy as np
import matplotlib.pyplot as plt

bright_red = [255, 0, 0]
dark_red = [106, 0, 0]
black = [0, 0, 0]
gray = [135, 135, 135]
dark_wood = [145, 65, 12]
bright_wood = [221, 137, 42]
green = [17, 139, 2]
```

```python
floor = np.zeros((8, 8, 3), dtype=np.uint8)
floor[:, :] = black

plt.imshow(floor)
plt.show()
```

```python
target = np.zeros((8, 8, 3), dtype=np.uint8)
target[0, :] = bright_red
target[7, :] = bright_red
target[:, 0] = bright_red
target[:, 7] = bright_red

# center 2x2 also red
target[3:5, 3:5] = bright_red

plt.imshow(target)
plt.show()
```

```python
box = np.zeros((8, 8, 3), dtype=np.uint8)
box[:, :] = bright_wood
box[0, :] = dark_wood
box[7, :] = dark_wood
box[:, 0] = dark_wood
box[:, 7] = dark_wood

box[np.arange(8), np.arange(8)] = dark_wood
box[np.arange(8), 7-np.arange(8)] = dark_wood

plt.imshow(box)
plt.show()
```

```python
# make a copy of box
box_on_target = np.copy(box)
box_on_target[0, :] = target[0, :]
box_on_target[7, :] = target[7, :]
box_on_target[:, 0] = target[:, 0]
box_on_target[:, 7] = target[:, 7]

box_on_target[1, 1] = bright_red
box_on_target[1, 6] = bright_red
box_on_target[6, 1] = bright_red
box_on_target[6, 6] = bright_red

plt.imshow(box_on_target)
plt.show()
```

```python
wall = np.zeros((8, 8, 3), dtype=np.uint8)
wall[:, :] = dark_red
wall[0, :] = gray
wall[4, :] = gray
wall[0:4, 0] = gray
wall[4:8, 4] = gray

plt.imshow(wall)
plt.show()
```

```python
player = np.zeros((8, 8, 3), dtype=np.uint8)
player[3:6, :] = green
player[6:8, 1] = green
player[6:8, 6] = green
player[2, 1] = green
player[2, 6] = green
player[2, 3:5] = green
player[1, 2:6] = green
player[0, 2] = green
player[0, 5] = green

plt.imshow(player)
plt.show()
```

```python
player_on_target = np.copy(target)

# put all green squares from player to player_on_target
player_on_target[player == green] = player[player == green]

plt.imshow(player_on_target)
plt.show()
```
